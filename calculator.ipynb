```python
def add(a,b):
    '''Return the sum of a and b
    >>> add(2,3)
    5
    >>> add(-1,1)
    0
    '''
    return a + b

def subtract(a,b):
    '''Return the difference of a and b
    >>> subtract(5,2)
    3
    >>> subtract(0,1)
    -1
    '''
    return a - b

print("Testing Person A's functions:")
print(f"add(2, 3) = {add(2, 3)}")
print(f"subtract(5, 2) = {subtract(5, 2)}")
print("Person A's functions work!")
```

```python
# Part 3: Multiplication (Person B)
def multiply(a, b):
    """Return a times b.
    >>> multiply(3, 4)
    12
    >>> multiply(0, 100)
    0
    """
    return a * b

# Part 4: Division (Person B)
def divide(a, b):
    """Return a divided by b. Raises ValueError if b is 0.
    >>> divide(10, 2)
    5.0
    >>> divide(7, 2)
    3.5
    """
    if b == 0:
        raise ValueError("Cannot divide by zero!")
    return a / b

# Test Person B's functions
print("Testing Person B's functions:")
print(f"multiply(3, 4) = {multiply(3, 4)}")
print(f"divide(10, 2) = {divide(10, 2)}")
print("âœ… Person B's functions work!")
```
